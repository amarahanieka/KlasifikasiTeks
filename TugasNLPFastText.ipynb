```python
import random
import numpy
import fasttext
import csv
import re
import pandas as pd
from sklearn.model_selection import train_test_split
import re
from gensim.utils import simple_preprocess
```

<h3>Membaca file dan melakukan simple prepsocessing </h3>

```python
data_test = pd.read_csv('data_worthcheck/train.csv')
print("any null values:", data_test.isnull().values.any())
```

```python
len(data_test)
```

```python
data_test.head(5)
```

```python
data_test = data_test.drop('Unnamed: 0', axis=1)
```

```python
data_test = data_test.rename(columns = {'text_a' : 'text', 'label' : 'label'})
```

```python
data_test.head(5)
```

```python
data_test.iloc[:, 0] = data_test.iloc[:, 0].apply(lambda x: ' '.join(simple_preprocess(x)))
```

```python
df_labeled = data_test.copy()
df_labeled['label'] = '__label__' + df_labeled['label'].astype(str)
df_labeled.head()
```

```python
label_prod = df_labeled['label'] + " " + df_labeled['text']
label_prod.head(5)
```

```python
data_test2 = pd.read_csv('data_worthcheck/test.csv')
print("any null values:", data_test2.isnull().values.any())
```

```python
data_test2 = data_test2.rename(columns = {'text_a' : 'text', 'label' : 'label'})
```

```python
data_test2.iloc[:, 0] = data_test2.iloc[:, 0].apply(lambda x: ' '.join(simple_preprocess(x)))
```

```python
df_labeled2 = data_test2.copy()
df_labeled2['label'] = '__label__' + df_labeled2['label'].astype(str)
df_labeled2.head()
```

```python
label_prod2 = df_labeled2['label'] + " " + df_labeled2['text']
```

<h3>Melakukan training dan testing supervised model dari fastText</h3>

```python
f_train = open("train.txt", "a")
for i in range(len(numpy.array(label_prod))):
    f_train.write(numpy.array(label_prod)[i] + "\n")
f_train.close()
```

```python
f_test = open("test2.txt", "a")
for i in range(len(numpy.array(label_prod2))):
    f_test.write(numpy.array(label_prod2)[i] + "\n")
f_test.close()
```

```python
model = fasttext.train_supervised(input="train.txt")
```

```python
model.test("test2.txt")
```

```python
model2 = fasttext.train_supervised(input="train.txt", wordNgrams=3, loss='hs', lr=0.5, epoch=25)
```

```python
model2.test("test2.txt")
```

```python
data_test2[['text', 'label']].to_csv('test2.txt', 
                                     index = False, 
                                     sep = ' ',
                                     header = None, 
                                     quoting = csv.QUOTE_NONE, 
                                     quotechar = "", 
                                     escapechar = " ")
```

<h2>Hasil prediksi model yang telah dilatih</h2>

```python
print("Hasil Prediksi model" + " | " + "Label yang seharusnya")
for i in range (len(data_test2)):
  print(str(model2.predict(data_test2.iloc[i, 0])) + " | " + str(data_test2.iloc[i, 1]))
```
